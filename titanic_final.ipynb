```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.drop(columns = ['Name','Ticket','Fare','Cabin','Embarked'],inplace=True)
```

```python
df.drop(columns = ['PassengerId'],inplace=True)
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.preprocessing import OneHotEncoder

# Initialize the encoder
encoder = OneHotEncoder()

# Apply fit_transform (make sure 'Sex' is reshaped or passed correctly)
en = encoder.fit_transform(df[['Sex']])
```

```python
en_array = en.toarray()
```

```python
en_array
```

```python
df.Sex = en_array[:,1]               #col. if male or not
```

```python
df.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(df.drop(columns=['Survived']),df.Survived)
```

```python
model.predict([[3,1,22,1,0]])
```
